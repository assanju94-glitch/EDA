```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/train_and_test2.csv")
```

```python
col = 'Age'
```

```python
data = df[col].dropna()
```

```python
plt.figure()
plt.hist(data)
plt.title(f"Histogram of {col}")
plt.xlabel(col)
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
plt.boxplot(data)
plt.title(f"Boxplot of {col}")
plt.show()
```

```python
plt.figure()
plt.scatter(range(len(data)), data)
plt.title(f"Scatter Plot of {col}")
plt.xlabel("Index")
plt.ylabel(col)
plt.show()
```

```python
sorted_data = data.sort_values().reset_index(drop=True)

plt.figure()
plt.plot(sorted_data)
plt.title(f"Line Chart of {col}")
plt.xlabel("Index")
plt.ylabel(col)
plt.show()
```

```python
top10 = data.sort_values(ascending=False).head(10)

plt.figure()
plt.bar(range(len(top10)), top10)
plt.title(f"Top 10 Highest {col}")
plt.xlabel("Top Entries")
plt.ylabel(col)
plt.show()
```

```python

top5 = data.sort_values(ascending=False).head(5)

plt.figure()
plt.pie(top5, labels=[f"Item {i}" for i in range(1,6)], autopct='%1.1f%%')
plt.title(f"Top 5 Contribution of {col}")
plt.show()
```

```python
num_df = df.select_dtypes(include='number')
```

```python
print("Numeric Columns:\n", num_df.columns)
```

```python
corr = num_df.corr()

print("\nCorrelation Matrix:\n", corr)
```

```python
plt.figure()

plt.imshow(corr)
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")
plt.show()
```
